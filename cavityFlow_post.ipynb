```python
import numpy as np
import matplotlib.pyplot as plt

# Loading data from text files
x = np.loadtxt('x.txt')
y = np.loadtxt('y.txt')
u = np.loadtxt('u.txt')
v = np.loadtxt('v.txt')
p = np.loadtxt('p.txt')

# Generating the meshgrid
X, Y = np.meshgrid(x, y)

# Plotting the fig
fig = plt.figure(figsize=(11, 7), dpi=100)

# Plotting the pressure field as a contour
plt.contourf(X, Y, p, alpha=0.5, cmap=plt.cm.viridis)
plt.colorbar()

# Plotting the pressure field outlines
plt.contour(X, Y, p, cmap=plt.cm.viridis)

# Plotting velocity field
plt.quiver(X[::2, ::2], Y[::2, ::2], u[::2, ::2], v[::2, ::2], scale=3, color='black')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Cavity Flow')

plt.show()
```
